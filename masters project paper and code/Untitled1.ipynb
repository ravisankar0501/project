```python
import numpy as np
import hashlib
import time
import secrets

def u(seed):
    # random number generator initialized with the seed
    np.random.seed(seed)
    return np.random.randint(0, 2**32)

def fingerprint_database(R, epsilon, k, Y):
    # Construct the fingerprintable set P
    n_records = len(R)
    n_attributes = len(R[0])
    P = []
    for i in range(n_records):
        for j in range(n_attributes):
            P.append((i, j))

    # Generate the internal ID
    IDinternal = Y.encode('utf-8')

    # Generate the fingerprint string
    h = hashlib.new('sha256')
    h.update(IDinternal)
    f = h.digest()

    # Generate the fingerprinted database
    M = R.copy()
    p = 1 / (np.exp(epsilon / k) + 1)
    L = len(f) * 8  # number of bits in the fingerprint
    cnt = 0
    start_time = time.time()
    for (i, j) in P:
        s = str(Y) + str(M[1, 0] + i + j)
        if u(s) % np.floor(1 / (2 * p)) == 0:
            x = u(s) % 2
            l = u(s) % L
            fi = (f[l//8] >> (l%8)) & 1  # extract the l-th bit of the fingerprint
            b = x ^ fi
            M[i, j] = b
            cnt += 1
    end_time = time.time()

    return R, M, cnt, end_time - start_time

import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv('adult.data')
df = df.drop(['1'], axis=1)

ohe = OneHotEncoder()
transformed = ohe.fit_transform(df[['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']])
print(fingerprint_database(transformed.toarray(), 0.2, 2, secrets.token_hex(32)))
```
