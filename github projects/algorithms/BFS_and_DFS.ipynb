```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque
```

```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=' ')

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
```

```python
def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)

    while queue:
        vertex = queue.popleft()
        print(vertex, end=' ')

        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
```

```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F', 'G'],
    'D': ['B'],
    'E': ['B', 'H'],
    'F': ['C'],
    'G': ['C'],
    'H': ['E']
}
```

```python
G = nx.Graph(graph)
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=20, font_weight='bold', \
        edge_color='black', linewidths=1, arrows=True)

plt.title('Graph Visualization')
plt.show()
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color='skyblue', font_size=20, font_weight='bold', \
        edge_color='black', linewidths=1, arrows=True)

plt.title('Graph Visualization')
plt.show()
```

```python
print("\nDepth-First Traversal:")
dfs(graph, 'A')  # Start DFS from vertex 'A'
```

```python
print("\nBreadth-First Traversal:")
bfs(graph, 'A')  # Start BFS from vertex 'A'
```
